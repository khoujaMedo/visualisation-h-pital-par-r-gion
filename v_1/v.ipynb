```python
import pandas as pd
import plotly.express as px

# Charger le dataset
try:
    df = pd.read_csv('dataset.csv')
    print("‚úÖ Dataset charg√© avec succ√®s!")
except:
    print("‚ùå Erreur de chargement. V√©rifiez le fichier dataset.csv")
    exit()

# L√©gende des cat√©gories
legend_categories = {
    "HP": "H√¥pital Provincial",
    "HR": "H√¥pital R√©gional", 
    "HIR": "Hospital Interr√©gional",
    "HPr": "H√¥pital de Proximit√©",
    "HPsyP": "H√¥pital Psychiatrique Provincial",
    "CRO": "Centre d'Oncologie",
    "HPsyR": "H√¥pital Psychiatrique R√©gional", 
    "CPU": "Centre Psychiatrique Universitaire"
}

# Pr√©parer les donn√©es
data = df.groupby(['region', 'categorie']).size().reset_index(name='count')
data['nom_categorie'] = data['categorie'].map(legend_categories)

print("üìä Donn√©es utilis√©es pour le treemap:")
print(data)

# Cr√©er le treemap
fig = px.treemap(
    data,
    path=['region', 'nom_categorie'],
    values='count',
    title='üè• H√¥pitaux du Maroc - Visualisation Treemap',
    color='count',
    color_continuous_scale='Blues'
)

# Personnaliser
fig.update_traces(
    textinfo="label+value+percent parent",
    hovertemplate="<b>%{label}</b><br>H√¥pitaux: %{value}<br>Part: %{percentParent:.1%}"
)

fig.show()
print("üéâ Treemap affich√©!")
```
